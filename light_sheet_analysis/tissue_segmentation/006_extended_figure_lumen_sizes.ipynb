```python
all_lumen_measurements = pd.DataFrame()
for position in tqdm(positions):
    output_dir_np_array = f"/cluster/project/treutlein/DATA/imaging/viventis/20220611_200804_ActGFP_LamRFP_TubRFFP_WTC_HB4_D4_pos1-4Mat_9-12_agar_processed/Position_{position}_Settings_1_Processed/organoid_measurements/"
    output_dir = f"/cluster/project/treutlein/DATA/imaging/viventis/20220611_200804_ActGFP_LamRFP_TubRFFP_WTC_HB4_D4_pos1-4Mat_9-12_agar_processed/Position_{position}_Settings_1_Processed/lumen_masks_v19_05_2023/"

    # stacks=np.linspace(1,125,10,dtype=int)
    for stack in tqdm(range(1, 125, 12)):
        # Get Organoid and lumen sizes
        combined_masks = imread(
            output_dir + f"{stack:04}" + "_lumen_organoid_mask_processed.tif"
        )
        lumen_mask = skimage.morphology.remove_small_objects(
            (combined_masks == 3), min_size=20000 / (2 * 1.388 * 1.388)
        )

        # Number of Lumen
        label_img = skimage.measure.label(lumen_mask)
        lumen_measurements = regionprops_table(
            label_img.astype(int), properties=("label", "area")
        )
        lumen_measurements = pd.DataFrame(lumen_measurements)
        lumen_measurements["hour"] = stack
        lumen_measurements["position"] = position
        all_lumen_measurements = all_lumen_measurements.append(
            lumen_measurements, ignore_index=True
        )

all_lumen_measurements[r"Log-volume of imaged lumen in $log(\mu m^3)$"] = np.log10(
    all_lumen_measurements["area"] * 0.347 * 4 * 0.347 * 4 * 2
)
all_lumen_measurements[r"Volume of imaged lumen in $\mu m^3*10^6$"] = (
    all_lumen_measurements["area"] * 0.347 * 4 * 0.347 * 4 * 2
) / 10**6
all_lumen_measurements["condition"] = all_lumen_measurements["position"]
all_lumen_measurements["condition"][all_lumen_measurements["position"] <= 4] = "ExtECM"
all_lumen_measurements["condition"][
    (all_lumen_measurements["position"] > 4) & (all_lumen_measurements["position"] <= 8)
] = "IntECM"
all_lumen_measurements["condition"][
    (all_lumen_measurements["position"] > 12)
] = "IntECM"
all_lumen_measurements["condition"][
    (all_lumen_measurements["position"] > 8)
    & (all_lumen_measurements["position"] <= 12)
] = "Diffusion barrier"
all_lumen_measurements["day"] = 4 + ((all_lumen_measurements["hour"] - 1) / 24)
```

```python
all_lumen_measurements_multimo = pd.DataFrame()
for position in tqdm(positions):
    output_dir = f"/cluster/project/treutlein/DATA/imaging/viventis/20210503_201032_6_lines_mosaic_HB4_D4_processed/Position_{position}_Settings_1_Processed/lumen_masks_v19_05_2023/"

    # stacks=np.linspace(1,125,10,dtype=int)
    for stack in tqdm(range(1, 249, 24)):
        combined_masks = imread(
            output_dir + f"{stack:04}" + "_lumen_organoid_mask_processed.tif"
        )
        lumen_mask = morphology.remove_small_objects(
            (combined_masks == 3), min_size=20000 / (2 * 1.388 * 1.388)
        )
        stack
        # Number of Lumen
        label_img = skimage.measure.label(lumen_mask)
        lumen_measurements = regionprops_table(
            label_img.astype(int), properties=("label", "area")
        )
        lumen_measurements = pd.DataFrame(lumen_measurements)
        lumen_measurements["hour"] = stack
        lumen_measurements["position"] = position
        all_lumen_measurements_multimo = all_lumen_measurements_multimo.append(
            lumen_measurements, ignore_index=True
        )

all_lumen_measurements_multimo[r"Log-volume of imaged lumen in $log(\mu m^3)$"] = (
    np.log10(all_lumen_measurements_multimo["area"] * 0.347 * 4 * 0.347 * 4 * 2)
)
all_lumen_measurements_multimo[r"Volume of imaged lumen in $\mu m^3*10^6$"] = (
    all_lumen_measurements_multimo["area"] * 0.347 * 4 * 0.347 * 4 * 2
) / 10**6
all_lumen_measurements_multimo["condition"] = "ExtECM"
all_lumen_measurements_multimo["day"] = 4 + (
    (all_lumen_measurements_multimo["hour"] - 1) / 48
)
```

```python
# Read tiffs + downscale
positions = [1, 2, 3, 4, 5, 6, 7, 8, 13, 14, 15, 16, 9, 10, 11, 12]
all_movies = []
for position in tqdm(positions):
    movie_rescaled = rescale(
        imread(f"movies/same_track_color_movies/tiff/thunder_plot_{position}.tiff"),
        [1, 0.5, 0.5, 1],
    )
    all_movies.append(movie_rescaled)
# Order
all_movies = np.array(all_movies)

montage_16 = []
for i in range(125):
    montage_16.append(
        skimage.util.montage(all_movies[:, i, 40:-40, 40:-40], channel_axis=-1)
    )
montage_16 = np.array(montage_16) * 255
imsave(
    f"movies/montage_16_thunderplots.tiff",
    np.moveaxis(montage_16.astype(np.uint8), -1, 1),
    imagej=True,
    metadata={"axes": "TCYX"},
    compression="zlib",
)
```

```python
# Read tiffs + downscale
positions = [1, 2, 3, 4, 5, 6, 7, 8, 13, 14, 15, 16, 9, 10, 11, 12]
all_movies = []
for position in tqdm(positions):
    movie_rescaled = rescale(
        imread(
            f"movies/90_percentile_slice/90_percentile_slice_positions_{position}_sep_channel.tiff"
        ),
        [1, 0.125, 0.125, 1],
        preserve_range=True,
    )
    all_movies.append(movie_rescaled)
# Pad
```

```python
array_shapes = np.array([image.shape for image in all_movies])
max_size = array_shapes.max(0)
padding = max_size - array_shapes
padded_movies_min = [
    np.pad(
        image,
        tuple([(pad[0], 0), (pad[1], 0), (pad[2], 0), (pad[3], 0)]),
        mode="constant",
        constant_values=0,
    )
    for image, pad in zip(all_movies, padding)
]
padded_movies_min = np.array(padded_movies_min)
```

```python
montage_16 = []
for i in range(125):
    montage_16.append(
        skimage.util.montage(padded_movies_min[:, i], channel_axis=-1, fill=[0, 0, 0])
    )
montage_16 = np.array(montage_16)
```

```python
imsave(
    f"movies/montage_90_percentile_slice.tiff",
    np.moveaxis(montage_16.astype(np.uint16), -1, 1),
    imagej=True,
    metadata={"axes": "TCYX"},
    compression="zlib",
)
```

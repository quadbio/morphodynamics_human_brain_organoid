```python
"""
BSD 3-Clause License

Copyright (c) 2021, DEXP
All rights reserved.

Redistribution and use in source and binary forms, with or without
modification, are permitted provided that the following conditions are met:

* Redistributions of source code must retain the above copyright notice, this
  list of conditions and the following disclaimer.

* Redistributions in binary form must reproduce the above copyright notice,
  this list of conditions and the following disclaimer in the documentation
  and/or other materials provided with the distribution.

* Neither the name of the copyright holder nor the names of its
  contributors may be used to endorse or promote products derived from
  this software without specific prior written permission.

THIS SOFTWARE IS PROVIDED BY THE COPYRIGHT HOLDERS AND CONTRIBUTORS "AS IS"
AND ANY EXPRESS OR IMPLIED WARRANTIES, INCLUDING, BUT NOT LIMITED TO, THE
IMPLIED WARRANTIES OF MERCHANTABILITY AND FITNESS FOR A PARTICULAR PURPOSE ARE
DISCLAIMED. IN NO EVENT SHALL THE COPYRIGHT HOLDER OR CONTRIBUTORS BE LIABLE
FOR ANY DIRECT, INDIRECT, INCIDENTAL, SPECIAL, EXEMPLARY, OR CONSEQUENTIAL
DAMAGES (INCLUDING, BUT NOT LIMITED TO, PROCUREMENT OF SUBSTITUTE GOODS OR
SERVICES; LOSS OF USE, DATA, OR PROFITS; OR BUSINESS INTERRUPTION) HOWEVER
CAUSED AND ON ANY THEORY OF LIABILITY, WHETHER IN CONTRACT, STRICT LIABILITY,
OR TORT (INCLUDING NEGLIGENCE OR OTHERWISE) ARISING IN ANY WAY OUT OF THE USE
OF THIS SOFTWARE, EVEN IF ADVISED OF THE POSSIBILITY OF SUCH DAMAGE.
"""
# Based on implementation from the projection function from DEXP: https://github.com/royerlab/dexp/blob/8e8399f5d0d8f1e1ae0ddfa6cb6011921929ae0b/dexp/processing/color/projection.py#L127


def attenuation_filter(
    image, attenuation_min_density, attenuation, attenuation_filtering
):
    if attenuation_filtering > 0:
        image_for_attenuation = gaussian_filter(image, sigma=attenuation_filtering)
    else:
        image_for_attenuation = image

    cum_density = cp.cumsum(
        attenuation_min_density + (1 - attenuation_min_density) * image_for_attenuation,
        axis=0,
    )

    image *= cp.exp(-attenuation * cum_density)
    return image
```

```python
import numpy as np
from matplotlib import pyplot as plt


def core_mask_creater(shapes, centers, radius, angles):
    center_x, center_y = centers
    shape_x, shape_y = shapes
    angle_1, angle_2 = angles
    radius = radius
    if angle_2 < angle_1:
        angle_2 += 360
    x, y = np.ogrid[-center_x : shape_x - center_x, -center_y : shape_y - center_y]
    mask = x * x + y * y <= radius * radius
    theta = np.rad2deg(np.arctan2(x, y))
    theta %= 360
    circle = np.zeros((shape_x, shape_y))
    circle[mask] = 1
    core_mask = theta <= (angle_2 - angle_1)
    return core_mask * circle
```

```python
# 1. multimosaic movie
positions = [10]
for position in positions:
    import os

    images_folder = f"/cluster/project/treutlein/DATA/imaging/viventis/20210503_201032_6_lines_mosaic_HB4_D4_processed/core_images_v17_06/"
    if not os.path.exists(images_folder):
        os.makedirs(images_folder)
    stack_file = h5py.File(
        f"/cluster/project/treutlein/DATA/imaging/viventis/20210503_201032_6_lines_mosaic_HB4_D4_processed/Position_{str(position)}_Settings_1_Processed/denoised_registered_processed.h5",
        "r",
    )
    cutout_wedge = [0.4, 0.4, 0.2, 0.2, 0.2, 0.2]
    cutout_wedge_angle = [30, 30, 30, 25, 20, 20]
    time_interval = 48
    all_mips = []
    for i in tqdm(range(6)):
        t = 1 + (time_interval) * i

        mempool = cp.get_default_memory_pool()
        pinned_mempool = cp.get_default_pinned_memory_pool()
        mempool.free_all_blocks()
        pinned_mempool.free_all_blocks()

        image = np.asarray(
            [
                stack_file[f"t{t:05}"]["s01"]["0"]["cells"][()].copy(),
                stack_file[f"t{t:05}"]["s00"]["0"]["cells"][()],
            ].copy()
        )

        image = image.astype(np.float32)

        down_scale_factor = 0.3
        image = ndimage.zoom(
            cp.asarray(image.astype(np.float32)),
            zoom=[
                1,
                down_scale_factor * (2 / 0.347),
                down_scale_factor * 1,
                down_scale_factor * 1,
            ],
            order=1,
        )

        mask = ~core_mask_creater(
            (image.shape[1], image.shape[3]),
            [int(475 * cutout_wedge[i]), int(689 * 0.5)],
            2000,
            (cutout_wedge_angle[i], 90),
        )
        cross_section_mask = np.zeros(image.shape[1:4])
        for i in range(image.shape[2]):
            cross_section_mask[:, i, :] = mask
        cross_section_mask = np.array(2 * [cross_section_mask])
        input_image = (cp.array(cross_section_mask) * image).get()
        angle = 160
        stack_mcherry_rotated = np.nan_to_num(
            rotate(
                cp.asarray(input_image[0]),
                angle,
                mode="constant",
                axes=(0, 2),
                reshape=True,
                order=0,
            )
        )
        print("rotated mcherry")
        stack_gfp_rotated = np.nan_to_num(
            rotate(
                cp.asarray(input_image[1]),
                angle,
                mode="constant",
                axes=(0, 2),
                reshape=True,
                order=0,
            )
        )
        print("rotated gfp")

        attenuation_filtering = 24
        attenuation_min_density = 0.3
        attenuation = 0.0005
        stack_mcherry_rotated = attenuation_filter(
            stack_mcherry_rotated,
            attenuation_min_density,
            attenuation,
            attenuation_filtering,
        ).get()
        print("attenuated mcherry")

        attenuation_filtering = 24
        attenuation_min_density = 0.3
        attenuation = 0.0005

        stack_gfp_rotated = attenuation_filter(
            stack_gfp_rotated,
            attenuation_min_density,
            attenuation,
            attenuation_filtering,
        ).get()
        print("attenuated gfp")

        plt.style.use("dark_background")
        # im_proj = np.max(np.flip(np.asarray([stack_mcherry_rotated,stack_gfp_rotated]), axis=3),axis = 1)
        # im_proj = skimage.exposure.rescale_intensity(im_proj, out_range=(0,1))
        # green_map = [[0,i/255,0] for i in range(256)]
        # green_matplotlib_map = ListedColormap(green_map, 'Green')
        # magenta_map = [[i/255,0,i/255] for i in range(256)]
        # magenta_matplotlib_map = ListedColormap(magenta_map, 'Magenta')

        # vmin_green, vmax_green = np.percentile(im_proj[0,:,:], q=(0.01, 99.9))
        """
        clipped_green = exposure.rescale_intensity(
            im_proj[0,:,:],
            in_range=(vmin_green, vmax_green),
            out_range=np.float32
        )

        vmin_magenta, vmax_magenta = np.percentile(im_proj[1,:,:], q=(0.01, 99.9))
        clipped_magenta = exposure.rescale_intensity(
            im_proj[1,:,:],
            in_range=(vmin_magenta, vmax_magenta),
            out_range=np.float32
        )

        channel1 = green_matplotlib_map(clipped_green)
        channel2 = magenta_matplotlib_map(clipped_magenta)
        assembled = np.stack((channel2, channel1), axis = 3)
        #rgb_proj = np.max(assembled, axis = 3)
        #fig, axes = plt.subplots(1,figsize = (16,16))
        #axes.imshow(rgb_proj)
        #axes.set_axis_off()
        #fig.tight_layout()
        #plt.show()
        """
        max_proj = np.max(
            np.flip(np.asarray([stack_mcherry_rotated, stack_gfp_rotated]), axis=3),
            axis=1,
        )
        channels = ["GFP", "mCherry"]
        all_mips.append(max_proj)

    all_mips = np.array(all_mips)
    print("saving")
    imsave(
        images_folder + f"position_{position}_{24}h_core_MIP.tif",
        all_mips.astype(np.uint16),
        imagej=True,
        resolution=(
            1.0 / (0.347 / down_scale_factor),
            1.0 / (0.347 / down_scale_factor),
        ),
        metadata={"unit": "um", "axes": "TCYX"},
        compression="zlib",
    )
```

```python
# 2. AGAR movie
positions = [15, 11]
# positions=[13,14,15,16]

all_mips = []
for position in positions:
    import os

    images_folder = f"/cluster/project/treutlein/DATA/imaging/viventis/20220611_200804_ActGFP_LamRFP_TubRFFP_WTC_HB4_D4_pos1-4Mat_9-12_agar_processed/core_images_v17_06/"
    if not os.path.exists(images_folder):
        os.makedirs(images_folder)
    stack_file = h5py.File(
        f"/cluster/project/treutlein/DATA/imaging/viventis/20220611_200804_ActGFP_LamRFP_TubRFFP_WTC_HB4_D4_pos1-4Mat_9-12_agar_processed/Position_{str(position)}_Settings_1_Processed/denoised_processed.h5",
        "r",
    )
    cutout_wedge = [0.2]
    cutout_wedge_angle = [20]
    time_interval = 24
    i = 0
    t = 1 + (time_interval) * 5

    mempool = cp.get_default_memory_pool()
    pinned_mempool = cp.get_default_pinned_memory_pool()
    mempool.free_all_blocks()
    pinned_mempool.free_all_blocks()

    image = np.asarray(
        [
            stack_file[f"t{t:05}"]["s01"]["0"]["cells"][()].copy(),
            stack_file[f"t{t:05}"]["s00"]["0"]["cells"][()],
        ].copy()
    )

    image = image.astype(np.float32)

    down_scale_factor = 0.3
    image = ndimage.zoom(
        cp.asarray(image.astype(np.float32)),
        zoom=[
            1,
            down_scale_factor * (2 / 0.347),
            down_scale_factor * 1,
            down_scale_factor * 1,
        ],
        order=1,
    )

    mask = ~core_mask_creater(
        (image.shape[1], image.shape[3]),
        [int(475 * cutout_wedge[i]), int(689 * 0.5)],
        2000,
        (cutout_wedge_angle[i], 90),
    )
    cross_section_mask = np.zeros(image.shape[1:4])
    for i in range(image.shape[2]):
        cross_section_mask[:, i, :] = mask
    cross_section_mask = np.array(2 * [cross_section_mask])
    input_image = (cp.array(cross_section_mask) * image).get()
    angle = 160
    stack_mcherry_rotated = np.nan_to_num(
        rotate(
            cp.asarray(input_image[0]),
            angle,
            mode="constant",
            axes=(0, 2),
            reshape=True,
            order=0,
        )
    )
    print("rotated mcherry")
    stack_gfp_rotated = np.nan_to_num(
        rotate(
            cp.asarray(input_image[1]),
            angle,
            mode="constant",
            axes=(0, 2),
            reshape=True,
            order=0,
        )
    )
    print("rotated gfp")

    attenuation_filtering = 24
    attenuation_min_density = 0.3
    attenuation = 0.0005
    stack_mcherry_rotated = attenuation_filter(
        stack_mcherry_rotated,
        attenuation_min_density,
        attenuation,
        attenuation_filtering,
    ).get()
    print("attenuated mcherry")

    attenuation_filtering = 24
    attenuation_min_density = 0.3
    attenuation = 0.0005

    stack_gfp_rotated = attenuation_filter(
        stack_gfp_rotated, attenuation_min_density, attenuation, attenuation_filtering
    ).get()
    print("attenuated gfp")

    plt.style.use("dark_background")
    im_proj = np.max(
        np.flip(np.asarray([stack_mcherry_rotated, stack_gfp_rotated]), axis=3), axis=1
    )
    im_proj = skimage.exposure.rescale_intensity(im_proj, out_range=(0, 1))
    green_map = [[0, i / 255, 0] for i in range(256)]
    green_matplotlib_map = ListedColormap(green_map, "Green")
    magenta_map = [[i / 255, 0, i / 255] for i in range(256)]
    magenta_matplotlib_map = ListedColormap(magenta_map, "Magenta")

    vmin_green, vmax_green = np.percentile(im_proj[0, :, :], q=(0.01, 99.9))
    clipped_green = exposure.rescale_intensity(
        im_proj[0, :, :], in_range=(vmin_green, vmax_green), out_range=np.float32
    )

    vmin_magenta, vmax_magenta = np.percentile(im_proj[1, :, :], q=(0.01, 99.9))
    clipped_magenta = exposure.rescale_intensity(
        im_proj[1, :, :], in_range=(vmin_magenta, vmax_magenta), out_range=np.float32
    )

    channel1 = green_matplotlib_map(clipped_green)
    channel2 = magenta_matplotlib_map(clipped_magenta)
    assembled = np.stack((channel2, channel1), axis=3)
    rgb_proj = np.max(assembled, axis=3)
    fig, axes = plt.subplots(1, figsize=(16, 16))
    axes.imshow(rgb_proj)
    axes.set_axis_off()
    fig.tight_layout()
    plt.show()
    max_proj = np.max(
        np.flip(np.asarray([stack_mcherry_rotated, stack_gfp_rotated]), axis=3), axis=1
    )
    channels = ["GFP", "mCherry"]
    all_mips.append(max_proj)


all_mips = np.array(all_mips)
print("saving")
```
